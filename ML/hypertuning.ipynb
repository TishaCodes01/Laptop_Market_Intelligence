```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('laptop_price_processed.csv')
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_col = ["Company", "TypeName", "OpSys", "Cpu_Brand", "Cpu_Series", "Gpu_Manufacturer", "Gpu_Series", "Memory_Type"]

le = LabelEncoder()

for col in categorical_col:
  df[col] = le.fit_transform(df[col])
```

```python
from sklearn.preprocessing import StandardScaler

numeric_col = ["Inches", "Ram", "Weight", "IPS_Panel", "Retina_Display", "Touchscreen", "Resolution_X", "Resolution_Y", "Clock_Speed", "Memory_Size(GB)"]

sc = StandardScaler()

for col in numeric_col:
  df[col] = sc.fit_transform(df[[col]])
```

```python
X = df.drop(columns=['laptop_ID', 'Price_euros', 'Product_Series', 'Total_Pixels'])
y = df['Price_euros']
```
