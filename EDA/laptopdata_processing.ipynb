```python
import numpy as np 
import pandas as pd 
```

```python
import chardet

# Detect encoding
with open('laptop_price.csv', 'rb') as file:
    result = chardet.detect(file.read())
    detected_encoding = result['encoding']
```

```python
detected_encoding
```

```python
dataset = pd.read_csv('laptop_price.csv', encoding=detected_encoding)
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.isna().sum()
```

```python
dataset['Ram'] = dataset['Ram'].str.replace('GB', '').astype(int)
dataset['Weight'] = dataset['Weight'].str.replace('kg', '').astype(float)
```

```python
dataset
```

```python
dataset['ScreenResolution'].unique()
```

```python
dataset['IPS_Panel'] = dataset['ScreenResolution'].apply(lambda x: 1 if 'IPS Panel' in x else 0)
dataset['Retina_Display'] = dataset['ScreenResolution'].apply(lambda x: 1 if 'Retina Display' in x else 0)
dataset['Touchscreen'] = dataset['ScreenResolution'].apply(lambda x: 1 if 'Touchscreen' in x else 0)
```

```python
dataset.info()
```

```python
resolution = dataset['ScreenResolution'].str.extract(r'(\d+)x(\d+)')
dataset['Resolution_X'] = resolution[0].astype(int)
dataset['Resolution_Y'] = resolution[1].astype(int)
```

```python
dataset['Total_Pixels'] = dataset['Resolution_X'] * dataset['Resolution_Y']
dataset['High_Resolution'] = dataset['Total_Pixels'].apply(lambda x: 1 if x > 1920*1080 else 0)
```

```python
dataset.info()
```

```python
dataset.drop('ScreenResolution', axis=1, inplace=True)
```

```python
dataset.info()
```

```python
dataset['Product'].unique()
```

```python
def product_series(product_str):
  if product_str.startswith('250') or product_str.startswith('255'):
    return product_str
  else:
    return product_str.split()[0]
```

```python
dataset['Product_Series'] = dataset['Product'].apply(product_series)
```

```python
dataset.drop('Product', axis=1, inplace=True)
```

```python
dataset['Cpu'].unique()
```

```python
dataset['Cpu_Brand'] = dataset['Cpu'].str.split().str[0]
dataset['Cpu_Series'] = dataset['Cpu'].str.extract(r'(Core i\d|Core M|Ryzen|Celeron|Pentium|Atom|Xeon|FX|E-Series|A\d{1,2}-Series|Cortex)')
dataset['Clock_Speed'] = dataset['Cpu'].str.extract(r'(\d+\.\d+|\d+)\s?GHz')
```

```python
dataset
```

```python
dataset['Cpu_Brand'].unique()
```

```python
dataset['Cpu_Series'].unique()
```

```python
dataset['Clock_Speed'].unique()
```

```python
dataset.drop('Cpu', axis=1, inplace =True)
```

```python
dataset['Gpu'].unique()
```

```python
dataset
```

```python
dataset['Gpu_Manufacturer'] = dataset['Gpu'].str.extract(r'(Nvidia|AMD|Intel|ARM)', expand=False)
```

```python
dataset['Gpu_Manufacturer'].unique()
```

```python
dataset['Gpu_Series'] = dataset['Gpu'].str.extract(r'(GeForce|Radeon|Iris Plus|Iris Pro|Iris|HD|UHD|Quadro|FirePro|Tesla|Vega|Mali)', expand=False)
```

```python
dataset['Gpu_Series'] = dataset['Gpu_Series'].fillna('Other')
```

```python
dataset['Gpu_Series'].unique()
```

```python
dataset
```

```python
dataset.drop('Gpu', axis=1, inplace=True)
```

```python
dataset['Memory'].unique()
```

```python
dataset['Memory_Type'] = dataset['Memory'].str.findall(r'\b(SSD|HDD|Flash Storage|Hybrid)\b').apply(lambda x: ' + '.join(set(x)))
```

```python
dataset['Memory_Type'].unique()
```

```python
dataset['Memory'] = dataset['Memory'].replace({
    r'\b1TB\b': '1024GB',
    r'\b1.0TB\b': '1024GB',
    r'\b2TB\b': '2048GB'
}, regex=True)
```

```python
dataset['Memory'].unique()
```

```python
def calculate_memory_size(memory_str):
    # Split by '+'
    components = memory_str.split('+')
    total_size = 0
    for component in components:
        # Extract numeric value (in GB)
        match = pd.Series(component).str.extract(r'(\d+)GB')
        if not match.empty:
            total_size += int(match[0][0])
    return total_size
```

```python
dataset['Memory_Size(GB)'] = dataset['Memory'].apply(calculate_memory_size)
```

```python
dataset.drop('Memory', axis=1, inplace=True)
```

```python
dataset.info()
```

```python
dataset.to_csv('laptop_price_processed.csv', index=False)
```

```python
column = ['High_Resolution', 'Touchscreen', 'Retina_Display', 'IPS_Panel']
plt.figure(figsize=(12, 4))
for i, col in enumerate(column):
  display = dataset[col].value_counts()
  plt.subplot(1, 4, i + 1) 
  display.plot(kind='bar')
  plt.title(f'Count for {col}')
plt.tight_layout() 
plt.show()
```

```python
column = ['High_Resolution', 'Touchscreen', 'Retina_Display', 'IPS_Panel']
plt.figure(figsize=(12, 4))
for i, col in enumerate(column):
  display_price = dataset.groupby(col)['Price_euros'].mean().reset_index()
  plt.subplot(1,4,i+1)
  sns.barplot(x=col, y='Price_euros', data=display_price, width=0.5)
  plt.title(f'Average Price for {col}')
plt.tight_layout()
plt.show()
```

```python
memory = dataset.groupby('Memory_Type')['Memory_Size(GB)'].mean().reset_index()
plt.figure(figsize=(12,5))
sns.barplot(x='Memory_Type', y='Memory_Size(GB)', data=memory, width=0.5)
plt.title('Average Memory Size by Memory Type')
plt.show()
```
